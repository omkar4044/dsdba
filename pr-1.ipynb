```python
 # Import required libraries 
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Load dataset from web source  
df=sns.load_dataset('titanic')
print(df.head())
```

```python
# Describe dataset (statistics, variables, dimensions)  
print(df.shape)
print(df.columns)
print(df.dtypes)
print(df.describe())
print(df.info())
print(df.nunique())
```

```python
# Handle missing values
print(df.isnull())
```

```python
print(df.isnull().sum())
```

```python
print(df.dropna(inplace=True))
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
# Fill missing 'age' with median
df['age'].fillna(df['age'].median())
```

```python
# Fill 'embarked' with mode
df['embarked'].fillna(df['embarked'].mode())
```

```python
# Drop 'deck' column (too many missing values)
df.drop(columns=['deck'])
```

```python
# Convert 'survived' to category
df['survived']=df['survived'].astype('category')
```

```python
df['survived'].dtypes
```

```python
# Convert 'pclass' to category
df['pclass']=df['pclass'].astype('category')
```

```python
df['pclass'].dtypes
```

```python
# Convert categorical variables to numerical  
df['sex']=df['sex'].map({'male':0,'female':1})
```

```python
df['sex'].dtypes
```
